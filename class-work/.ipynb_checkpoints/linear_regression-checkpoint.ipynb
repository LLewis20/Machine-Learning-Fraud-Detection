```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split

# bring in the credit card fraud dataset
data_frame = pd.read_csv('FraudDataset\\fraudTrain.csv')


# the dataset on kaggle has a train and test set already split but here is how I would do it for the dataset I'm using.
train_set, test_set = train_test_split(data_frame, test_size = 0.2)

# making a working copy
working_set = train_set.copy()
```

```python
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error

plt.scatter(working_set['amt'], working_set['is_fraud'])
plt.xlabel("Amount")
plt.ylabel("is fraud")
plt.show()
```

* I picked amt because it had the largest correlation coefficient with .02 which is low but it has the highest.

* is_fraud is what we are trying to find

```python
reg = LinearRegression()
X = working_set[['amt']]
Y = working_set['is_fraud']
reg.fit(X,Y)

print("The bias is " , reg.intercept_)
print("The feature coefficients are ", reg.coef_)
```

## Plotting the line

```python
theta0 = reg.intercept_
theta1 = reg.coef_[0]
print(theta0, theta1)
def the_line(h, bias, coef):
    return bias + coef * h

min_height = working_set["amt"].min()
min_weight = the_line(min_height, theta0, theta1)
max_height = working_set["amt"].max()
max_weight = the_line(max_height, theta0, theta1)
line_x = [min_height, max_height]
line_y = [min_weight, max_weight]


import matplotlib.pyplot as plt
plt.scatter(working_set["amt"], working_set["is_fraud"])
plt.xlabel("amt")
plt.ylabel("is_fraud")
plt.plot(line_x, line_y)
plt.show()
```

## Measure the model


```python
print(reg.score(X,Y))
```

## Mean Squared Error

```python
y_pred = reg.predict(X)
mse = mean_squared_error(Y, y_pred)
print("mean squared error is ",  mse)

from math import sqrt
rmse = sqrt(mse)
print("The root mean squared error is ", rmse)
```

* The score for this model is only return a 5% which is very low
